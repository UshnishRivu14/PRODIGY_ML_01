# IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

**IMPORTING THE HOUSE PRICES DATA** 

```python
df = pd.read_csv('/kaggle/input/home-prices/train.csv')
df.head()
```

**SETTING X AND Y VALUES**

```python
features = ['GrLivArea','BedroomAbvGr','FullBath']
```

```python
x = df[features]
y = df['SalePrice']
```

```python
x
```

```python
y
```

**SPLITING THE DATA INTO TRAINING DATA AND TESTING DATA**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state =25)
```

```python
x_train.shape
```

```python
x_test.shape
```

**IMPORTING LINEAR REGRESSION MODEL**

```python
linear_model = LinearRegression()
```

**TRAINING THE MODEL**

```python
linear_model.fit(x_train,y_train)
```

**PREDICTING FOR TESTING DATASET**

```python
y_pred = linear_model.predict(x_test)
```

**CALCULATING MEAN SQUARED ERROR**

```python
mse = mean_squared_error(y_test, y_pred)
```

**CALCUTING R2 SCORE**

```python
r2 = r2_score(y_test, y_pred)
```

**CALCULATING MEAN SQUARED ERROR AND R2 SCORE FOR EVALUATING MODEL PERFORMANCE**

```python
print("Mean Squared Error: ",mse)
print("r2 score: ",r2)
```

**PLOTING THE ACTUAL PRICE VS PREDICTED PRICE**

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price VS Predicted Price')
plt.show()
```

**PREDICTING FOR RANDOM VALUES**

```python
new_data = pd.DataFrame({'GrLivArea':[2000],'BedroomAbvGr':[3],'FullBath':[2]})
predicted_price = linear_model.predict(new_data)
print(f"Predicted Price: {predicted_price[0]}")
```
